```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = 'D:/Projects/grip/SampleSuperstore.csv'
df = pd.read_csv(data_path)
```

Dropping the unimportant columns

```python
data=df.drop(['Postal Code','Country'],axis=1)
```

```python
data.columns
```

<b>Profit based on category<b/>

```python
fig, axes = plt.subplots(figsize=(15, 4))
sns.barplot(data,x="Sub-Category",y="Sales", errorbar=None)
plt.show()
fig, axes = plt.subplots(figsize=(15, 4))
sns.barplot(data,x="Sub-Category",y="Discount",color="red", errorbar=None)
plt.show()
fig, axes = plt.subplots(figsize=(15, 4))
sns.barplot(data,x="Sub-Category",y="Profit",color="green",errorbar=None)
plt.show()
```

<b>Region wise sales & profit based on category<b/>

```python
fig, axes = plt.subplots(1, 3,figsize=(8, 4))  # 1 row, 2 columns

# Plot the first barplot on the first axis
sns.barplot(data=data, x="Region", y="Sales", hue="Category", ax=axes[0], errorbar=None)
axes[0].set_title('Sales by Region')

sns.barplot(data=data, x="Region", y="Discount", hue="Category", ax=axes[1], errorbar=None)
axes[1].set_title('Discount by Region')
# Plot the second barplot on the second axis
sns.barplot(data=data, x="Region", y="Profit", hue="Category", ax=axes[2], errorbar=None)
axes[2].set_title('Profit by Region')


plt.tight_layout()
# Show the plots
plt.show()
```

Here all regions have equal sales but discount in central region was too high comapred to other regions.
This leads to reduced profit in central region.

<b>Region wise sales & profit based on category<b/>

```python
sns.barplot(data,x="Category",y="Profit",hue="Region", errorbar=None)
plt.show()
sns.barplot(data,x="Category",y="Discount",hue="Region", errorbar=None)
plt.show()
```

From the above the charts we can conclude that the region or categories with high discount has very less profits

<b> Analysing state-wise sales, profit</b>

```python
# state wise profit
fig, axes = plt.subplots(1, 3,figsize=(15, 7))  # 1 row, 2 columns

sns.barplot(data,y="State",x="Profit",orient = 'h', ax=axes[2], errorbar=None)
axes[2].set_title('Profit by States')
sns.barplot(data,y="State",x="Sales",orient = 'h', ax=axes[0], errorbar=None)
axes[0].set_title('Sales by States')

sns.barplot(data,y="State",x="Discount",orient = 'h', ax=axes[1], errorbar=None)
axes[1].set_title('Discount by States')
plt.tight_layout()
# Show the plots
plt.show()
```


Here, the state illinois,Texas and Pennsylavania has highest discount rates which caused them loss.


The states with nearly zero discount, has high profit rates compared to other district.

Finally by analyzing category-wise, region and state-wise, we can conclude that discount rate is the major factor affecting the profit.

So, by reducing discount rates we can increase the profit.
